Câu 1 Lập trình đối tượng và tính toán số học

phần 1 ý a

```python
#xây dựng lớp cha
class NhanVien:
    def __init__(self, ma_nv: str, luong_cb: float):
        self.__ma_nv = ma_nv
        self.__luong_cb = luong_cb

    def tinh_luong(self):
        return self.__luong_cb

    def get_ma_nv(self):
        return self.__ma_nv
```

phần 2 ý a

```python
#xây dựng lớp con
class NhanVienTangCa(NhanVien):
    def tinh_luong(self):
        return super().tinh_luong() * 1.2


class NhanVienChuyenChuyen(NhanVien):
    def tinh_luong(self):
        return super().tinh_luong() + 500_000
```

phần 3 ý a

```python
# chương trình chính(main)
ds_nv = [
    NhanVien("NV01", 10_000_000),
    NhanVienTangCa("NV02", 10_000_000),
    NhanVienChuyenChuyen("NV03", 10_000_000)
]

for nv in ds_nv:
    print(f"Mã NV: {nv.get_ma_nv()} | Lương: {nv.tinh_luong():,.0f} VND")
```

Phần 1 ý b

```python
#Chuyển danh sách thành mảng NumPy và reshape
import numpy as np

san_luong = [12, None, 8, 15, None, 10]

arr = np.array(san_luong, dtype=float)
arr = arr.reshape(2, 3)
```

Phần 2 ý b

```python
# Thay NaN bằng giá trị trung bình
mean_value = np.nanmean(arr)
arr = np.nan_to_num(arr, nan=mean_value)

print("Mảng sau khi xử lý:")
print(arr)
```

Phần 3 ý b

```python
#Trích xuất phần tử > 9 và tìm giá trị nhỏ nhất
filtered = arr[arr > 9]
min_value = np.min(filtered)

print("Các phần tử > 9:", filtered)
print("Giá trị nhỏ nhất:", min_value)
```

```python
import pandas as pd
import sqlite3

# a) Đọc dữ liệu
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("nhan_su.csv")

# Lọc Ma_NV chứa 'NS' (không phân biệt hoa thường, xử lý NaN)
# và Nam_Cong_Tac >= 2021
df_filtered = df[
    df["Ma_NV"].astype(str).str.contains("NS", case=False, na=False)
    & (df["Nam_Cong_Tac"] >= 2021)
]


# b) Điền giá trị thiếu Thanh_Tich bằng trung bình (sau lọc)
mean_thanh_tich = df_filtered["Thanh_Tich"].mean()
df_filtered["Thanh_Tich"] = df_filtered["Thanh_Tich"].fillna(mean_thanh_tich)

# c) Groupby – Tổng Thanh_Tich theo Phong_Ban
grouped = df_filtered.groupby("Phong_Ban")["Thanh_Tich"].sum()
print("Tổng Thanh_Tich theo Phòng Ban:")
print(grouped)

# d) Pivot Table
pivot_table = pd.pivot_table(
    df_filtered,
    index="Phong_Ban",
    columns="Nam_Cong_Tac",
    values="Thanh_Tich",
    aggfunc="sum",
    fill_value=0
)

print("\nPivot Table:")
print(pivot_table)

# e) Lưu vào SQLite và truy vấn
conn = sqlite3.connect("HeThong.db")

pivot_table.to_sql(
    "bang_luong_2024",
    conn,
    if_exists="replace"
)

# Truy vấn 5 dòng đầu tiên
query = "SELECT * FROM bang_luong_2024 LIMIT 5"
result = pd.read_sql(query, conn)

print("\n5 dòng đầu tiên trong bảng bang_luong_2024:")
print(result)

conn.close()
```
